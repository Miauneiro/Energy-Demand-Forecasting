```python
# test_apis.py
import requests
import os
from pathlib import Path

# Read .env file manually (simple version)
env_path = Path('.env')
if env_path.exists():
    with open(env_path) as f:
        for line in f:
            if '=' in line:
                key, value = line.strip().split('=', 1)
                os.environ[key] = value

# Test EIA
eia_key = os.environ.get('EIA_API_KEY')
print(f"Testing EIA API...")
eia_url = f"https://api.eia.gov/v2/electricity/rto/region-data/data/?api_key={eia_key}&frequency=hourly&data[0]=value&facets[respondent][]=US48&start=2024-11-01T00&end=2024-11-02T00"
eia_response = requests.get(eia_url)
print(f"EIA Status: {eia_response.status_code}")
if eia_response.status_code == 200:
    print("✅ EIA API works!")
else:
    print(f"❌ EIA failed: {eia_response.text}")

print()

# Test OpenWeather
weather_key = os.environ.get('OPENWEATHER_API_KEY')
print(f"Testing OpenWeather API...")
weather_url = f"http://api.openweathermap.org/data/2.5/weather?q=NewYork&appid={weather_key}"
weather_response = requests.get(weather_url)
print(f"Weather Status: {weather_response.status_code}")
if weather_response.status_code == 200:
    print("✅ OpenWeather API works!")
else:
    print(f"❌ Weather failed: {weather_response.text}")
```
